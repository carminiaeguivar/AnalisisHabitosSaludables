```python
demo = pd.read_csv("demographic_n.csv")
diet = pd.read_csv("diet.csv")      
exam = pd.read_csv("examination.csv")
labs = pd.read_csv("labs.csv")
meds = pd.read_csv("medications.csv", encoding = 'latin-1')
#locat = pd.read_csv("Locations.csv" ,encoding = 'latin-1')
ques = pd.read_csv("questionnaire.csv")
```

```python
#alldata = join_all(list(demo,diet,exam,labs,meds,ques));
#Combine diet questions

#demo.drop(['SEQN'], axis = 1, inplace=True)
diet.drop(['SEQN'], axis = 1, inplace=True)
exam.drop(['SEQN'], axis = 1, inplace=True)
labs.drop(['SEQN'], axis = 1, inplace=True)
meds.drop(['SEQN'], axis = 1, inplace=True)
ques.drop(['SEQN'], axis = 1, inplace=True)

alldata = pd.concat([demo, diet], axis=1, join='inner')
alldata = pd.concat([alldata, exam], axis=1, join='inner')
alldata = pd.concat([alldata, labs], axis=1, join='inner')
alldata = pd.concat([alldata, meds], axis=1, join='inner')
alldata = pd.concat([alldata, ques], axis=1, join='inner')
```
